```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('data.csv')
df['lua'] = df['lua'].astype('category')  # Optimize for categorical
```

```python
# Convert energy to Joules
df['energy_joules'] = df['energy_consumed'] * 3.6e6

plt.figure(figsize=(16, 9))
sns.boxplot(
    data=df,
    x='lua',
    y='energy_joules',
    color='#2ecc71',          # Matching green color
    linewidth=1.5,
    flierprops=dict(
        markerfacecolor='#27ae60',
        markeredgecolor='#27ae60',
        markersize=5
    )
)

plt.title('', pad=20, fontsize=14)
plt.ylabel('Energy Consumed (J)', fontsize=12, labelpad=15)
plt.xlabel('', fontsize=12, labelpad=15)
plt.xticks(rotation=45, ha='right')

# Format y-axis and remove scientific notation
plt.ticklabel_format(axis='y', style='plain', useOffset=False)

# Add light grid lines
plt.grid(True, axis='y', alpha=0.3, linestyle='--')
plt.grid(True, axis='x', alpha=0.3, linestyle='--')

# Remove plot borders
sns.despine(left=True, bottom=True)

# Adjust plot margins
plt.tight_layout()
plt.show()
```

```python
# Calculate average energy consumption and convert to Joules
duration = df.groupby(['lua'])['duration'].mean().reset_index()

# Create line plot
plt.figure(figsize=(14, 7))
sns.lineplot(
    data=duration,
    x='lua',
    y='duration',
    marker='o',         # Add circular markers
    markersize=10,
    linewidth=2.5,
    color="#82ddf9",    # A pleasant green color
    markeredgecolor="#82ddf9",
    markerfacecolor='white'
)

plt.title('', pad=20, fontsize=14)
plt.xlabel('', labelpad=15, fontsize=12)
plt.ylabel('Runtime (s)', labelpad=15, fontsize=12)
plt.xticks(rotation=45, ha='right')

# Format y-axis to show normal numbers instead of scientific notation
plt.ticklabel_format(axis='y', style='plain', useOffset=False)

# Add grid lines
plt.grid(True, alpha=0.3, linestyle='--')

# Add data point labels
for index, row in duration.iterrows():
    plt.annotate(
        f'{row["duration"]:.2f} s',
        (row['lua'], row['duration']),
        textcoords="offset points",
        xytext=(0, 10),
        ha='center',
        fontsize=10,
        fontweight='bold',
        color='#2c3e50'
    )

# Remove borders
sns.despine(left=True, bottom=True)

plt.tight_layout()
plt.show()
```

```python
# Convert energy to milliwatt-hours and calculate averages
df['energy_mwh'] = df['energy_consumed'] * 1e6
avg_energy = df.groupby(['lua', 'benchmark'])['energy_mwh'].mean().reset_index()

plt.figure(figsize=(14, 7))
sns.barplot(
    data=avg_energy,
    x='lua',
    y='energy_mwh',
    hue='benchmark',
    # palette='Greens_d',
    edgecolor='white',
    linewidth=1,
)

plt.title('Average Energy Consumption by Version and Benchmark', pad=20, fontsize=14)
plt.ylabel('Average Energy Consumed (mWh)', fontsize=12, labelpad=15)
plt.xlabel('Lua Version', fontsize=12, labelpad=15)
plt.xticks(rotation=45, ha='right')

# Format y-axis and remove scientific notation
plt.ticklabel_format(axis='y', style='plain', useOffset=False)

# Style the legend
plt.legend(
    title='Benchmark',
    bbox_to_anchor=(1.05, 1),
    loc='upper left',
    frameon=False,
    edgecolor='white'
)

# Add grid lines
plt.grid(True, axis='y', alpha=0.3, linestyle='--')
plt.grid(True, axis='x', alpha=0.3, linestyle='--')

# Remove borders
sns.despine(left=True, bottom=True)

plt.tight_layout()
plt.show()
```
